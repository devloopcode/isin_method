```python
import pandas as pd
import numpy as np

df_laptops = pd.read_csv('laptop_price.csv', encoding='unicode_escape')
```

```python
df_laptops.head()
```

# 1. isin(): Single filter

```python
# select Apple or Dell
df_laptops[df_laptops['Company'].isin(['Apple', 'Dell'])].head()
```

```python
 df_laptops[df_laptops['Company'].isin(['Apple', 'Dell'])].value_counts('Company')
```

# 2. isin(): Multiple filtering

```python
# find Ultrabook or Notebook from Apple OR Dell
filter_one = df_laptops['TypeName'].isin(['Ultrabook', 'Notebook'])
filter_two = df_laptops['Company'].isin(['Apple', 'Dell'])
```

```python
df_laptops[filter_one & filter_two].head()
```
